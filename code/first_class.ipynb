```python
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
import sys

# 이미지 업로드
uploaded = files.upload()

uploaded = cv2.imread('lettuce.jpg')

if uploaded is None:
    print('Image load failed!')
    sys.exit()

cv2_imshow(uploaded)
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# 이미지 전처리
image = cv2.cvtColor(uploaded, cv2.COLOR_BGR2RGB)
image = image.reshape((image.shape[0] * image.shape[1], 3))  # height, width 통합



k = 2
clt = KMeans(n_clusters=k)
clt.fit(image)

def centroid_histogram(clt):
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins=numLabels)
    hist = hist.astype("float")
    hist /= hist.sum()
    return hist

hist = centroid_histogram(clt)

def plot_colors(hist, centroids):
    bar = np.zeros((50, 300, 3), dtype="uint8")
    startX = 0
    for (percent, color) in zip(hist, centroids):
        endX = startX + (percent * 300)
        cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
                      color.astype("uint8").tolist(), -1)
        startX = endX
    return bar

bar = plot_colors(hist, clt.cluster_centers_)

plt.figure(figsize=(8, 4))
plt.axis("off")
plt.imshow(bar)
plt.show()
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 이미지 전처리
image = cv2.cvtColor(uploaded, cv2.COLOR_BGR2RGB)
image = image.reshape((image.shape[0] * image.shape[1], 3))  # height, width 통합
image = np.delete(image, np.where(np.all(image > [245, 245, 245], axis=1)), axis=0)

k = 2
clt = KMeans(n_clusters=k)
clt.fit(image)

def centroid_histogram(clt):
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins=numLabels)
    hist = hist.astype("float")
    hist /= hist.sum()
    return hist

hist = centroid_histogram(clt)

def plot_colors(hist, centroids):
    bar = np.zeros((50, 300, 3), dtype="uint8")
    startX = 0
    for (percent, color) in zip(hist, centroids):
        endX = startX + (percent * 300)
        cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
                      color.astype("uint8").tolist(), -1)
        startX = endX
    return bar

bar = plot_colors(hist, clt.cluster_centers_)

plt.figure(figsize=(8, 4))
plt.axis("off")
plt.imshow(bar)
plt.show()
```

```python
# 필요한 모듈 임포트
from yellowbrick.cluster import SilhouetteVisualizer

# k값 범위 설정
k_range = range(2, 5)

# 각 k값에 대한 시각화 및 실루엣 지수 출력
for k in k_range:
    clt = KMeans(n_clusters=k)
    visualizer = SilhouetteVisualizer(clt, colors='yellowbrick')
    visualizer.fit(image)
    print(f"K = {k}, Silhouette Score: {visualizer.silhouette_score_: .4f}")
    visualizer.show()
```
